## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
words_upper = []
for word in words:
    words_upper.append(word.upper())    
words_upper 
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
words_leng = []
for word in words:
    if len(word) > 5:
        words_leng.append(word)
words_leng
```

**Print the first word starting with "t"**

```python
# your code here
first_word_t = []
for word in words:
    if word.startswith('t'):
        first_word_t.append(word)   
first_word_t
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# your code here
square_numbers = []
for number in range(1, 11):
    square_numbers.append(number ** 2)  
square_numbers
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here
odd_numbers = []
for number in range(1, 11):
    if number % 2 != 0:
        odd_numbers.append(number)
odd_numbers
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# your code here
multiples_of_8  = []
for number in range(1, 101):
    if number % 8 == 0:
        multiples_of_8.append(number)
multiples_of_8
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# your code here
people_total = len(people)
people_total
```

**How many people have kids**?

```python
# your code here
people_with_kids = 0
for person in people:
    if person["n_kids"] > 0:
        people_with_kids += 1
people_with_kids
```

**How many kids do they have in total?**

```python
# your code here
total_kids = 0
for person in people:
    total_kids += person["n_kids"]
total_kids
```
